```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,root_mean_squared_error,mean_squared_error
```

```python
df=pd.read_csv("/Users/prathamsharma/Desktop/ML/Social_Network_Ads.csv")
df.head()
```

```python
sns.boxplot(df["Age"])
```

```python
sns.boxplot(df["EstimatedSalary"])
```

```python
stats.probplot(df["EstimatedSalary"],dist="norm",plot=plt)
plt.show()
```

```python
x=df.iloc[:,3:4]
x.head()
```

```python
y=df.iloc[:,-1]
y.head(5)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
print(x_test.shape)
print(x_train.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
print("True label:", y_train.iloc[0])
print("Predicted label:", lr.predict(x_train.iloc[[0]])[0])
```
